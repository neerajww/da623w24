## Taking Fourier spectrum of images
Let's first create an analog of 1-D sine wave signal in 2-D. These 2-D sine waves are also referred to as grating signals. 

```python
# below creates a 1-D sine wave signal in spatial x-direction
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-500, 501, 1)
wavelength = 200 # the wavelength parameter here is analog to time period paramter for time-domain signals 
y = np.sin(2 * np.pi * x / wavelength)
plt.plot(x, y)
plt.xlabel('x [distance]')
plt.ylabel('y(x)')
plt.show()
```

```python
# below creates a 2-D sine wave also referred to as a grating signal
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-500, 501, 1)
X, Y = np.meshgrid(x, x)
wavelength = 200
grating = np.sin(2 * np.pi * X / wavelength)

fig = plt.subplots(figsize=[9,9])
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, grating, 50, cmap='binary')
ax.set_title('Grating Signal - A 2-D sine wave')
plt.show()

fig = plt.subplots(figsize=[9,9])
plt.set_cmap("gray")
plt.imshow(grating)
plt.title('A 2-D projection of the grating - 2-D sine wave')
plt.show()

```

```python
# adding an orientation to the grating
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-500, 501, 1)
X, Y = np.meshgrid(x, x)
wavelength = 200
angle = np.pi / 4
grating = np.sin(2*np.pi*(X*np.cos(angle) + Y*np.sin(angle)) / wavelength)
plt.set_cmap("gray")
plt.imshow(grating)
plt.show()
```

## Taking 2-D Fourier transform of gratings 

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-500, 501, 1)
X, Y = np.meshgrid(x, x)
wavelength = 200
angle = 0
grating = np.sin(
    2*np.pi*(X*np.cos(angle) + Y*np.sin(angle)) / wavelength
)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=[12,4])
FS = 12

plt.set_cmap("gray")
ax[0].imshow(grating)
# Calculate Fourier transform of grating
# centering the image for odd number of columns/rows
ft = np.fft.ifftshift(grating)
nfft = 1024
ft = np.fft.fft2(grating, [nfft, nfft])
# ft = np.fft.fft2(grating)
# centering the 2-D fft
ft = np.fft.fftshift(ft)

ax[1].imshow(abs(ft))
plt.xlabel('$f_x$')
plt.ylabel('$f_y$')
plt.xlim([nfft//2 - 20, nfft//2 + 20])
plt.ylim([nfft//2 + 20, nfft//2  - 20])

fx_indx = np.arange(nfft//2 - 20, nfft//2 + 20, 5)
fx_freq = (fx_indx-nfft//2) / nfft / 1
ax[1].set_xticks(fx_indx)
ax[1].set_xticklabels(fx_freq, rotation=90)

ax[1].set_yticks(fx_indx)
ax[1].set_yticklabels(fx_freq, rotation=0)
plt.show()
```

## Computing inverse 2-D DFT of the image spectrum

```python
# Calculate the inverse Fourier transform of 
# the Fourier transform
ift = np.fft.fftshift(ft)
ift = np.fft.ifft2(ift)
ift = ift.real  # Take only the real part
# plt.subplot(133)
plt.imshow(ift)
plt.show()
```

## Operating on natural image

```python
im = plt.imread('taj_mahal.jpeg')
im_gs = im.mean(axis=2)

grating = im_gs.copy()
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=[16,6])
FS = 12

plt.set_cmap("gray")
ax[0].imshow(grating)
# Calculate Fourier transform of grating
# centering the image for odd number of columns/rows
ft = np.fft.fftshift(grating)
nfft = im_gs.shape[0]
ft = np.fft.fft2(grating, [nfft, nfft])
# ft = np.fft.fft2(grating)
# centering the 2-D fft
ft = np.fft.fftshift(ft)

ax[1].imshow(10*np.log(abs(ft)))
plt.xlabel('$f_x$')
plt.ylabel('$f_y$')
plt.xlim([0, nfft])
plt.ylim([nfft, 0])
plt.show()
```

## Invert the 2-D FT

```python
# Calculate the inverse Fourier transform of 
# the Fourier transform
ift = np.fft.ifftshift(ft)
ift = np.fft.ifft2(ift)
ift = ift.real  # Take only the real part
# plt.subplot(133)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=[6,6])
plt.imshow(ift)
plt.show()
```

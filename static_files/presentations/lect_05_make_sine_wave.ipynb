```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
```

This is an example code to create and visualize sine waves and polynomials.

```python
# define domain
dx = 0.0001
cycles = 2
L = np.pi
x = np.arange(-cycles*L+dx, cycles*L+dx, dx)
n = len(x)
```

```python
# make sin waves
m = 1

f = np.cos(m * np.pi * x/L)
fig, ax = plt.subplots(figsize=[9,5])
# set color map for sequence of overlaid traces
name = "Accent"
cmap = get_cmap('tab10')
colors = cmap.colors
ax.set_prop_cycle(color=colors)

LW = 2
FS  = 14
ax.plot(x, f)
ax.set_xlabel('x (domain variable)',fontsize=14)
ax.set_ylabel('f(x)',fontsize=FS) # A.U stands for Arbitrary Units
ax.grid(color='gray', linestyle='--', linewidth=1,alpha=.3)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)  
```

```python
# make cosine waves

fig, ax = plt.subplots(figsize=[4,8])
# set color map for sequence of overlaid traces
name = "Accent"
cmap = get_cmap('tab10')
colors = cmap.colors
ax.set_prop_cycle(color=colors)

LW = 2
FS  = 14

for m in range(1,10):
    x = np.arange(0,np.pi/8, dx)
    f = np.cos(m * np.pi * x/L)
    ax.plot(x, f)

ax.set_xlabel('x (domain variable)',fontsize=14)
ax.set_ylabel('f(x)',fontsize=FS) # A.U stands for Arbitrary Units
ax.grid(color='gray', linestyle='--', linewidth=1,alpha=.3)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```

```python
# polynomials and sine waves

fig, ax = plt.subplots(figsize=[4,4])
# set color map for sequence of overlaid traces
name = "Accent"
cmap = get_cmap('tab10')
colors = cmap.colors
ax.set_prop_cycle(color=colors)

LW = 2
FS  = 14

for m in range(1,5):
    x = np.arange(0,np.pi, dx)
    f_sin = np.sin(m * np.pi * x/L + np.random.rand(1)*np.pi)
    ax.plot(x, f_sin)

ax.set_xlabel('x (domain variable)',fontsize=14)
ax.set_ylabel('f(x)',fontsize=FS) # A.U stands for Arbitrary Units
ax.grid(color='gray', linestyle='--', linewidth=1,alpha=.3)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)  

fig, ax = plt.subplots(figsize=[4,4])
# set color map for sequence of overlaid traces
name = "Accent"
cmap = get_cmap('tab10')
colors = cmap.colors
ax.set_prop_cycle(color=colors)

LW = 2
FS  = 14

for m in range(1,5):
    x = np.arange(0,np.pi, dx)
    f_poly = x**m
    ax.plot(x, f_poly)

ax.set_xlabel('x (domain variable)',fontsize=14)
ax.set_ylabel('f(x)',fontsize=FS) # A.U stands for Arbitrary Units
ax.grid(color='gray', linestyle='--', linewidth=1,alpha=.3)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)  
```
